## Скрипт генерации сетчатки

Скрипт генерирует базовую версию сетчатки в соответствии с параметрами, прописанными в `config.ini`.

```python
from Retina import Retina

retina = Retina('config.ini')
retina.createPoints()
retina.plot()
```

Далее сетчатка поворачивается перпендикулярно нужному нам вектору и декартовы координаты всех ее точек записываются в файл.

```python
retina.rotateTo([150., 100., 50.])
print("Writing output to the rotated retina file...")
retina.saveToFile()
print("Done.")
retina.plot()
```
